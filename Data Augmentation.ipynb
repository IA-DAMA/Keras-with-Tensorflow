# Data Augmentation

**Data augmentation happens when you create new data based on modifications, besides, Data augmentation om images would include transformations like:**

    * Flipping image either horizontally or vertically
    * Rotating the image
    * Zooming in or out on the image
    * Cropping the image
    * Varying the color of the image

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20, 20))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1,
       height_shift_range=0.1, shear_range=0.15, zoom_range=0.1,
       channel_shift_range=10, horizontal_flip=True)
```

```python
chosen_image = random.choice(os.listdir(r'C:\Machine_Learning\dogs-vs-cats\train\train\dog'))
```

```python
image_path = 'C:\\Machine_Learning\\dogs-vs-cats\\train\\train\\dog\\' + chosen_image
```

```python
assert os.path.isfile(image_path)
```

```python
#Obtain image
image = np.expand_dims(plt.imread(image_path),0)
plt.imshow(image[0])
```

```python
#Generate batches of augmented images from this image
aug_iter = gen.flow(image)
```

```python
#Get 10 samples of augmented images
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
#Augmented images
plotImages(aug_images)
```
